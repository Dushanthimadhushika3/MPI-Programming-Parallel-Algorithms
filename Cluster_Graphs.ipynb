```python
#Merge sort sequential performance
import matplotlib.pyplot as plt
   
Time = [0.186,0.483,0.649,1.304,1.283,1.411,1.903,2.141,3.005,2.739,4.094,3.527,3.938,5.362,4.996,5.988,5.538,4.604,6.179,7.158]
No_of_elements = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000,15000,16000,17000,18000,19000,20000]
  
plt.figure(figsize=(10,6))
plt.plot(No_of_elements, Time,color='green', marker='o')
plt.title('Merge Sort - Sequential')
plt.ylabel('Execution Time (ms)')
plt.xlabel('No. of elements')
plt.grid(True)
plt.show()
```

```python
#Merge sort parallel performance with 2 processors
Time = [0.000456,0.000733,0.001093,0.014227,0.001953,0.017940,0.002027,0.002314,0.010080,0.004039,0.012028,0.008642,0.004770,0.011225,0.009334,0.010626,0.015292,0.005185,0.016386,0.011859]
No_of_elements = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000,15000,16000,17000,18000,19000,20000]

Final = []
for i in Time:
  i = i*1000
  Final.append(i)

plt.figure(figsize=(10,6))
plt.plot(No_of_elements, Final,color='green', marker='o')
plt.title('Merge Sort - Parallel (2 Processors)')
plt.ylabel('Execution Time (ms)')
plt.xlabel('No. of elements')
plt.grid(True)
plt.show()
```
